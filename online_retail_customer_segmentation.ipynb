```python
import pandas as pd
from datetime import datetime ,timedelta,date
import plotly.express as px
```

```python
df=pd.read_excel('Online Retail.xlsx')
```

```python
def get_run_date():
    base = datetime.today()
    ojo_oni = base.strftime("%Y-%m-%d")
    ojo_oni=pd.to_datetime(ojo_oni)
    return ojo_oni
```

```python
df=df[df['Quantity']>0] 
```

```python
df.info()
```

```python
df['CustomerID']=df['CustomerID'].astype(str)
```

```python
df.info()
```

```python
df['Total_amount']=df['Quantity']* df['UnitPrice']
```

```python
dff=df.groupby('CustomerID').agg(max_tran_date=('InvoiceDate', 'max'),
                            monetory_value=('Total_amount', 'sum'),
                             frequency=('Total_amount', 'count')).reset_index()
```

```python
dff
```

```python
dff['recency']=get_run_date()-dff['max_tran_date']
dff['recency']=dff['recency'].astype(str)
```

```python
dff['Recency']=dff['recency'].str.extract(r'(\d{4})')
```

```python
dff['Recency']=dff['Recency'].astype(int)-4137
```

```python
dff
```

```python
menetory_list=list(dff['monetory_value'].quantile([0.25,0.5,0.75,0.9]))
menetory_list
```

```python
frequency_list=list(dff['frequency'].quantile([0.25,0.5,0.75,0.9]))
frequency_list
```

```python
recency_list=list(dff['Recency'].quantile([0.25,0.5,0.75,0.9]))
recency_list
```

```python
recency_list=list(dff['Recency'].quantile([0.25,0.5,0.75,0.9]))
recency_list
```

```python
def recency(score):
    if score<=recency_list[0]:
        x=5
    elif score>recency_list[0] and score <=recency_list[1]:
        x=4
    elif score>recency_list[1] and score <=recency_list[2]:
        x=3
    elif score>recency_list[2] and score <=recency_list[3]:
        x=2
    elif score>recency_list[3]:
        x=1
    return x
```

```python
def frequency(score):
    if score<=frequency_list[0]:
        x=1
    elif score>frequency_list[0] and score <=frequency_list[1]:
        x=2
    elif score>frequency_list[1] and score <=frequency_list[2]:
        x=3
    elif score>frequency_list[2] and score <=recency_list[3]:
        x=4
    elif score>recency_list[3]:
        x=5
    return x
```

```python
def monetary_value(score):
    if score<=menetory_list[0]:
        x=1
    elif score>menetory_list[0] and score <=menetory_list[1]:
        x=2
    elif score>menetory_list[1] and score <=menetory_list[2]:
        x=3
    elif score>menetory_list[2] and score <=menetory_list[3]:
        x=4
    elif score>menetory_list[3]:
        x=5
    return x
```

```python
dff['m_score']=dff['monetory_value'].apply(lambda x:monetary_value(x))
dff['r_score']=dff['Recency'].apply(lambda x:recency(x))
dff['f_score']=dff['frequency'].apply(lambda x:frequency(x))
```

```python
dff['rfm_score']=dff['m_score']+dff['r_score']+dff['f_score']
```

```python
dff['r_score']=dff['Recency'].apply(lambda x:recency(x))
```

```python
rfm_score=list(dff['sum_total'].quantile([0.25,0.5,0.75,0.9]))
```

```python
def final_score(score):
    if score<=rfm_score[0]:
        x=1
    elif score>rfm_score[0] and score <=rfm_score[1]:
        x=2
    elif score>rfm_score[1] and score <=rfm_score[2]:
        x=3
    elif score>rfm_score[2] and score <=rfm_score[3]:
        x=4
    elif score>rfm_score[3]:
        x=5
    return x
```

```python
dff['rfm_final_score']=dff['sum_total'].apply(lambda x:final_score(x))
```
